```python
# ----------- preprocessing section -----------#
# 1- handle null value 
df['TypeOfBook'].fillna(value="مجهول",inplace=True)
df['author'].fillna(value="مجهول",inplace=True)
df['PublishingHouse'].fillna(value="مجهول",inplace=True)
df['BookSummary'].fillna(value=" ",inplace=True)
```

```python
corpus =np.array( df['title']+' '+df['author']+' '+df['BookSummary']+' '+df['PublishingHouse']+' '+df['TypeOfBook'])
# corpus = np.array(df['title'])
print(corpus[3])
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.isri import ISRIStemmer
import re # Regular expression library
import string
# import nltk
# nltk.download('punkt')
# import nltk
# nltk.download('stopwords')
document_tokens =[]
def remove_stop_words(tokens):
    result_tokens =[]
    stopwords_list = stopwords.words('arabic')
    for token in tokens:
        if token not in stopwords_list:
            result_tokens.append(token)
    return result_tokens
# make stemming for arabic word using IRIStemmer
def stemming(tokens):
    st = ISRIStemmer()
    result_tokens=[]
    for token in tokens:
        result_tokens.append(st.stem(token))
    return result_tokens
for i in range(len(corpus[0:1000])):
    # Remove Punctuation from text
    corpus[i] = re.sub('[%s]' % re.escape(string.punctuation), '', str(corpus[i]))
    corpus[i] = re.sub('،','',corpus[i])
    corpus[i] = re.sub('“','',corpus[i])
    corpus[i] = re.sub(r'\s*[A-Za-z]+\b','',corpus[i])
    # Remove any digit number from text 
    corpus[i] = re.sub('\d+','',corpus[i])
    corpus[i] = re.sub('PDF','',corpus[i])
    corpus[i] = re.sub('كتب','',corpus[i])
    corpus[i] = re.sub('كتاب','',corpus[i])
    corpus[i] = re.sub('الكاتب','',corpus[i])
    
    # convert text to tokinze 
    tokens = word_tokenize(str(corpus[i]))
    # Remove Stop word from text 
    process_Tokens = remove_stop_words(tokens)
    stemming_Tokens =stemming(process_Tokens)
    document_tokens.append(stemming_Tokens)
#     # convert tokens to one text 
    corpus[i] =' '.join(stemming_Tokens)
#     df['title'][i]=' '.join(stemming_Tokens_title)

print(str(document_tokens[0]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import operator
## Create Vocabulary
vocabulary = set()
for doc in list_token:
    vocabulary.update(doc.split(','))
vocabulary = list(vocabulary)
# Intializating the tfIdf model
tfidf = TfidfVectorizer(vocabulary=vocabulary)
# Fit the TfIdf model
tfidf.fit(list_token)
# Transform the TfIdf model
tfidf_tran=tfidf.transform(list_token)
featurs_name = tfidf.get_feature_names()
```

```python
def gen_vector_T(tokens):
  Q = np.zeros((len(vocabulary)))    
  x= tfidf.transform(tokens)
  #print(tokens[0].split(','))
  for token in tokens[0].split(','):
      #print(token)
      try:
          ind = vocabulary.index(token)
          Q[ind]  = x[0, tfidf.vocabulary_[token]]
      except:
          pass
  return Q
```

```python
def cosine_sim(a, b):
    cos_sim = np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))
    return cos_sim
```

```python
def cosine_similarity_T(k, query):
    # preprocessed_query = re.sub("\W+", " ", query).strip()
    tokens = word_tokenize(str(query))
    # Remove Stop word from text 
    process_Tokens = remove_stop_words(tokens)
    stemming_Tokens =stemming(process_Tokens)
    string =''
    list_token=[]
    for token in stemming_Tokens:
      string = string + ',' +token
    # list_token.append(string)
    # document_tokens.append(stemming_Tokens)
    # tokens = word_tokenize(str(preprocessed_query))
    q_df = pd.DataFrame(columns=['q_clean'])
    q_df.loc[0,'q_clean'] =str(string)
    
    # q_df['q_clean'] =wordLemmatizer(q_df.q_clean)
    d_cosines = []
    
    query_vector = gen_vector_T(q_df['q_clean'])
    for d in tfidf_tran.A:
        d_cosines.append(cosine_sim(query_vector, d))
                  
    out = np.array(d_cosines).argsort()[-k:][::-1]
    #print("")
    d_cosines.sort()
    a = pd.DataFrame()
    for i,index in enumerate(out):
        a.loc[i,'index'] = str(index)
        a.loc[i,'title'] = df['title'][index]
        a.loc[i,'author']= df['author'][index]

        a.loc[i,'BookSummary']=df['BookSummary'][index]
    for j,simScore in enumerate(d_cosines[-k:][::-1]):
        a.loc[j,'Score'] = simScore
    return a
```

```python
# rank=cosine_similarity_T(5, ' تهذيب المهلكات والمنجزات ')
rank=cosine_similarity_T(5,  'كتاب رحلة في الكون')
# print(rank['title'][0:10])
rank
# print(df['title'])
```

```python
corpus_doc= rank['title']+' '+ rank['author']+' '+ rank['BookSummary']
 # convert text to tokinze 
list_of_tokens=[]
for i in range(len(corpus_doc)):
  tokens = word_tokenize(str(corpus_doc[i]))
  # Remove Stop word from text 
  process_Tokens = remove_stop_words(tokens)
  stemming_Tokens =stemming(process_Tokens)
  list_of_tokens.append(' '.join(stemming_Tokens))
```

```python
def extract_topn_from_vector(feature_names, sorted_items, topn=10):
    """get the feature names and tf-idf score of top n items"""
    
    #use only topn items from vector
    sorted_items = sorted_items[:topn]

    score_vals = []
    feature_vals = []
    
    # word index and corresponding tf-idf score
    for idx, score in sorted_items:
        
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    #create a tuples of feature, score
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]
    
    return results
def sort_coo(coo_matrix):
    """Sort a dict with highest score"""
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

def get_keywords(vectorizer, feature_names, doc):
    """Return top k keywords from a doc using TF-IDF method"""

    #generate tf-idf for the given document
    tf_idf_vector = vectorizer.transform([doc])
    
    #sort the tf-idf vectors by descending order of scores
    sorted_items=sort_coo(tf_idf_vector.tocoo())
    TOP_K_KEYWORDS=15
    #extract only TOP_K_KEYWORDS
    keywords=extract_topn_from_vector(feature_names,sorted_items,TOP_K_KEYWORDS)
    
    return list(keywords.keys())

result = []
i=0
for doc1 in list_of_tokens:
    df_word = {}
    df_word['full_text'] = corpus_doc[i]
    df_word['top_keywords'] = get_keywords(tfidf, featurs_name, doc1)
    result.append(df_word)
    i+=1
    
final = pd.DataFrame(result)
final
```
